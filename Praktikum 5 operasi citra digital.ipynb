1. Operasi Negatif

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
inp = cv2.imread('data/zero.png')
cv2.imshow('Gambar original', inp)
plt.imshow(inp)
plt.title("Gambar ori BGR")
plt.show()

inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)
plt.imshow(inpRGB)
plt.title("Gambar ori RGB")
plt.show()

tombol = cv2.waitKey(0) & 0xFF
cv2.destroyAllWindows()
```

```python
d = inp.shape
r = d[0]
c = d[1]
out = inp

for x in range(0, r):
    for y in range(0, c):
        red = 255 - inp[x, y, 0]
        green = 255 - inp[x, y, 1]
        blue = 255 - inp[x, y, 2]
        out[x, y] = [red, green, blue]

cv2.imshow('Gambar negatif', out)
outRgb = cv2.cvtColor(out, cv2.COLOR_BGR2RGB)
plt.imshow(outRgb)
plt.title("Gambar Negatif")
plt.show()
tombole = cv2.waitKey(0)

if tombol == ord('s'):
    cv2.imwrite('gamabarNegatif.png', out)

cv2.destroyAllWindows()
```

2. Operasi Thershold

```python
imgGray = cv2.cvtColor(inpRGB, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap="gray")
```

```python
d = imgGray.shape
r = d[0]
c = d[1]
out = np.zeros(d)
T = 120

for x in range(0, r):
    for y in range(0, c):
        if(imgGray[x, y] < T):
            out[x, y] = 0
        else:
            out[x, y] = 255

plt.imshow(out, cmap="gray")
plt.title("Gambar Thershold")
plt.show()

tombol =  cv2.waitKey(0) & 0xFF
if tombol == ord('s'):
    cv2.imwrite('gambarthershold.png', out)
cv2.destroyAllWindows()
```

2.3. Operasi Clipping

```python
def oprClipping(img):
    d = img.shape
    r = d[0]
    c = d[1]
    out = img

    for x in range(0, r):
        for y in range(0, c):
            if (img[x, y] > 255):
                out[x, y] = 255
            elif (img[x, y] < 0):
                out[x, y]
            else:
                out[x, y] = img[x, y]

    return out

if __name__ == "__main__":
    inp = cv2.imread('data/bentukwarna.png')
    inpGray = cv2.cvtColor(inp, cv2.COLOR_BGR2GRAY)
    plt.title("Gambar 2 Dimensi")
    plt.imshow(inpGray, cmap="gray")
    plt.show()
    plt.title("Gambar Hasil Clipping")
    plt.imshow(oprClipping(inpGray), cmap="gray")
    plt.show()
    
```

```python
def oprNeg(img):
    d = img.shape
    r = d[0]
    c = d[1]
    out = img

    for x in range(0, r):
        for y in range(0, c):
            for l in range(0, 3):
                out[x, y, l] = 255 - img[x, y, l]
    
    outRGB = cv2.cvtColor(out, cv2.COLOR_BGR2RGB)
    plt.title("Harusnya gambar negatif")
    plt.subplot(121)
    plt.imshow(outRGB)
    plt.show()

if __name__ == "__main__":
    inp = cv2.imread('data/zero.png')
    inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)
    plt.title("Harusnya gambar original")
    plt.subplot(121)
    plt.imshow(inpRGB)
    plt.show()
    oprNeg(inp)
    cv2.destroyAllWindows()
```

Tugas 2: buatlah program dari operasi thershold

```python
def oprTh(img):
    d = img.shape
    r = d[0]
    c = d[1]
    out = np.zeros(d)
    T = 140

    for x in range(0, r):
        for y in range(0, c):
            if(img[x, y] < T):
                out[x, y] = 0
            else:
                out[x, y] = 255

    cv2.imshow('Gambar Thershold', out)

    tombol =  cv2.waitKey(0) & 0xFF
    if tombol == ord('s'):
        cv2.imwrite('gambarThershold.png', out)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    inp = cv2.imread('data/zero.png')
    inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)
    inpGray = cv2.cvtColor(inpRGB, cv2.COLOR_BGR2GRAY)
    cv2.imshow('gambar gray', inpGray)
    oprTh(inpGray)
    cv2.destroyAllWindows()
```

Tugas 3: Buatlah algoritma berupa pseudocode dan program dalam bahasa python!

Pseudocode

```python
def oprBr(img) :
    d = img.shape
    r = d[0]
    c = d[1]
    out = img
    b = 2

    for x in range(1, r):
        for y in range(1, c):
            out[x, y] = img[x, y] * b
    return out

inp = cv2.imread('data/my-dream.jpg')
inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)

plt.subplot(121), plt.imshow(inpRGB), plt.title("Original")
plt.subplot(122), plt.imshow(oprBr(inpRGB)), plt.title("Pencahayaan")

```

Tugas 4: Buatlah algoritma berupa pseudocode dan program dalam Bahasa Python!

Pseudocode

```python
f = cv2.imread('data/kameramen.png')
f = cv2.cvtColor(f, cv2.COLOR_BGR2GRAY)
g = cv2.imread('data/beras.png')
g = cv2.cvtColor(g, cv2.COLOR_BGR2GRAY)

def clipping(img1, img2):
    d = f.shape
    r = d[0]
    c = d[1]
    h = img1
    for x in range(r):
        for y in range(c):
            if img1[x, y] > 255:
                img1[x, y] = 255
            elif img1[x, y] < 0:
                img1[x, y] = 0
            if img2[x, y] > 255:
                img2[x, y] = 255
            elif img2[x, y] < 0:
                img2[x, y] = 0
            else:
                img1[x, y] = img1[x, y]
                img2[x, y] = img2[x, y]
            
            h[x, y] = img1[x, y] - img2[x, y]

    return h

plt.subplot(131), plt.imshow(f, cmap="gray"), plt.title("Citra 1")
plt.subplot(132), plt.imshow(g, cmap="gray"), plt.title("Citra 2")
plt.subplot(133), plt.imshow(clipping(f,g), cmap="gray"), plt.title("Citra Gabungan")
```

Tugas 6: Buatlah algortima berupa pseudocode dan program dalam Bahasa python!

Translation

```python
tx = 50
ty = 20

a = cv2.imread('data/beras.png')
a = cv2.cvtColor(a, cv2.COLOR_RGB2GRAY)

d = a.shape
r = d[0]
c = d[1]

m = np.float32([[1, 0, tx], [0, 1, ty]])

h = cv2.warpAffine(a, m, (r, c))

plt.subplot(121), plt.title('Original'), plt.imshow(a, cmap="gray")
plt.subplot(122), plt.title('Hasil'), plt.imshow(h, cmap="gray")
```
