*1 Color Mapping*

```python
import cv2
import matplotlib.pyplot as plt;
import numpy as np;
```

```python
img = cv2.imread('data/color_space.jpg')
plt.imshow(img)
plt.title("Original Image")
```

```python
cv2.imshow("Citra Asli", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

*1.1 RGB*

```python
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
red = imgrgb[:,:,0]
plt.subplot(131)
plt.imshow(red)
plt.title('Layer Red')
plt.subplot(132)
plt.imshow(imgrgb[:,:,1])
plt.title('Layer Green')
plt.subplot(133)
plt.imshow(imgrgb[:,:,2])
plt.title('Layer Blue')
```

```python
plt.subplot(131)
plt.imshow(imghsv[:,:,0])
plt.title('Layer Hue')
plt.subplot(132)
plt.imshow(imghsv[:,:,1])
plt.title("Layer Sturation")
plt.subplot(133)
plt.imshow(imghsv[:,:,2])
plt.title("Layer Value")
```

*1.7 Implementasi mencari warna*

```python
img = cv2.imread('data/bentukwarna.png')
if not img is None:
    # Konversi ke HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    batasRendah = np.array([100, 20, 20])
    batasAtas = np.array([140, 255, 255])

    # Peroleh bagian yang berwarna biru
    getBiru = cv2.inRange(hsv, batasRendah, batasAtas)

    # Kenakan operasi and terhadap citra asli
    hasil = cv2.bitwise_and(img, img, mask = getBiru)

    cv2.imshow('Citra asli', img)
    cv2.imshow('Masking', getBiru)

    cv2.imshow('Citra asli', hasil)
    cv2.waitKey()
    cv2.destroyAllWindows()
else:
    print("Citra tidak ditemukan")
```

```python
img = cv2.imread('data/bentukwarna.png')
if not img is None:
    # Konversi ke HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    batasRendahA = np.array([159, 20, 20])
    batasAtasA = np.array([179, 255, 255])
    
    batasRendahB = np.array([0, 20, 20])
    batasAtasB = np.array([20, 255, 255])

    # Peroleh bagian yang berwarna merah
    maskA = cv2.inRange(hsv, batasRendahA, batasAtasA)
    maskB = cv2.inRange(hsv, batasRendahB, batasAtasB)
    masking = maskA | maskB

    # Kenakan operasi and terhadap citra asli
    hasil = cv2.bitwise_and(img, img, mask = masking)

    cv2.imshow('Citra asli', img)
    cv2.imshow('Masking Matriks', masking)

    cv2.imshow('Citra asli', hasil)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Citra tidak ditemukan !")
```
