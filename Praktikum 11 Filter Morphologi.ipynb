```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

1 Filter Morphologi

1.1 Fungsi untuk load gambar (create)

```python
def load_img(x="2003071"):
    blank_img = np.zeros((900,900))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text=x, org=(50,300), fontFace=font, fontScale=5, color=(255,255,255), thickness=25, lineType=cv2.LINE_AA)
    return blank_img
```

1.2 Fungsi untuk menampilkan gambar

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    plt.show()
```

```python
img = load_img()
cv2.imshow('gambar', img)
display_img(img)

tombol = cv2.waitKey(0) & 0xFF
# print(ord(tombol))
if tombol == ord('s'):
    cv2.imwrite('coba.png', img)
cv2.destroyAllWindows()
```

1.3 Filter Erosi

```python
kernel = np.ones((5,5), np.uint8)
erosion1 = cv2.erode(img, kernel, iterations=1)
display_img(erosion1)
img = load_img()
kernel = np.ones((5,5), np.uint8)
erosion5 = cv2.erode(img, kernel,iterations=4)
display_img(erosion5)
```

1.4 Filter Dilasi

```python
dilation = cv2.dilate(erosion5, kernel, iterations=1)
display_img(dilation)
dilation5 = cv2.dilate(dilation, kernel, iterations=1)
display_img(dilation5)
```

1.5 Filter Opening

```python
img = load_img()
white_noise = np.random.randint(low=0, high=2,size=(900,900))
white_noise = white_noise*255

white_noise.shape
img.shape
noise_img = white_noise + img
display_img(noise_img)
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
display_img(opening)
```

1.6 Filter Closing

```python
img = load_img()
black_noise = np.random.randint(low=0, high=2, size=(900,900))
black_noise = black_noise * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img==-255] = 0
display_img(black_noise_img)
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_img(closing)
```

1.7 Morphological Gradient

```python
img = load_img()
display_img(img)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
